```python
import matplotlib.pyplot as plt
import numpy as np
import tifffile
%matplotlib inline
```

```python
fig, axarr = plt.subplots(1, 3, figsize=(48,16))
ax1 = axarr[0]
ax2 = axarr[1]
ax3 = axarr[2]

names = ["beadpack", "ketton", "estaillades"]#, "Beadpack",
         #"Estaillades", "Bentheimer", "Doddington"]
direc = "image_data/images/"

images = []
for name in names:
    im = tifffile.imread(direc+name+".tif")
    images.append(im[0, :,  :])
    
for ax, name, image, x_label, y_label in zip(axarr, names,
                                             images,
                                             [r'$x-Axis \ (Voxels)$', r'$x-Axis \ (Voxels)$', r'$x-Axis \ (Voxels)$'],
                                             [r'$y-Axis \ (Voxels)$', r'$y-Axis \ (Voxels)$', r'$y-Axis \ (Voxels)$']):

    ax.set_title(r'$'+name.title()+'$', size=45, y=1.02)


    ax.imshow(image, cmap='Greys',  interpolation='nearest')
    for item in ax.get_xticklabels():
        item.set_fontsize(28)
    for item in ax.get_yticklabels():
        item.set_fontsize(28)

    ax.set_xlabel(x_label, size=40)
    ax.set_ylabel(y_label, size=40)
    ax.set_xlim(0, im.shape[0]+1)
    ax.set_ylim(0, im.shape[1]+1)
fig.canvas.draw()

plt.show()
fig.savefig("image_data/image_views/image_views_results_section.png", bbox_extra_artists=None, bbox_inches='tight', dpi=300)
plt.close()
```
