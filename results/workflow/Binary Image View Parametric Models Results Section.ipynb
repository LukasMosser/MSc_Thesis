```python
import matplotlib.pyplot as plt
import numpy as np
import tifffile
from matplotlib.patches import Circle, Ellipse, Arrow
%matplotlib inline
```

```python
fig, axarr = plt.subplots(1, 3, figsize=(48,16))
ax1 = axarr[0]
ax2 = axarr[1]
ax3 = axarr[2]

names = ["Isotropic14", "Anisotropic131314", "Anisotropic121214"]#, "Beadpack",
         #"Estaillades", "Bentheimer", "Doddington"]
direc = "image_data/images/"

images = []
for name in names:
    im = tifffile.imread(direc+name+".tif")
    images.append(im[:, 1,  :])
    
for ax, name, title, image, x_label, y_label in zip(axarr, names,[r'$Isotropic \ r_{1,2,3}=1.4$',
                                                                 r'$Anisotropic \ r_{1,2}=1.3 \ r_{3}=1.4$',
                                                                 r'$Anisotropic \ r_{1,2}=1.2 \ r_{3}=1.4$'],
                                             images,
                                             [r'$x-Axis \ (Voxels)$', r'$x-Axis \ (Voxels)$', r'$x-Axis \ (Voxels)$'],
                                             [r'$z-Axis \ (Voxels)$', r'$z-Axis \ (Voxels)$', r'$z-Axis \ (Voxels)$']):

    ax.set_title(title, size=45, y=1.02)
    if name == "Isotropic14":
        ax.add_artist(Circle((99, 400,), 68, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Circle((198, 400,), 68, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Circle((198, 300,), 68, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Circle((99, 300,), 68, linestyle='--', linewidth=2, color='w', fill=False))
        ax.annotate("",xy=(247, 447), xycoords='data', xytext=(200, 400), textcoords='data',
                    arrowprops=dict(arrowstyle="->", fc="w", ec="w", linewidth=2),)
        
        ax.text(0.430, 0.865, r"$r_{1,2,3}$", ha='center', va='center', 
                transform=ax.transAxes, color="w", fontsize=38, rotation=45)
    elif name == "Anisotropic131314":
        ax.add_artist(Ellipse((99, 400,), 130, 140, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Ellipse((198, 400,), 130, 140, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Ellipse((198, 300,), 130, 140, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Ellipse((99, 300,), 130, 140, linestyle='--', linewidth=2, color='w', fill=False))

        ax.annotate("",xy=(265, 400), xycoords='data', xytext=(200, 400), textcoords='data',
                    arrowprops=dict(arrowstyle="->", fc="w", ec="w", linewidth=2))
        ax.annotate("",xy=(200, 470), xycoords='data', xytext=(200, 400), textcoords='data',
                    arrowprops=dict(arrowstyle="->", fc="w", ec="w", linewidth=2))
        
        ax.text(0.47, 0.825, r"$r_{1,2}$", ha='center', va='center', 
                transform=ax.transAxes, color="w", fontsize=38, rotation=0)
        ax.text(0.43, 0.9, r"$r_{3}$", ha='center', va='center', 
                transform=ax.transAxes, color="w", fontsize=38, rotation=0)   
    elif name == "Anisotropic121214":
        ax.add_artist(Ellipse((99, 400,), 120, 140, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Ellipse((198, 400,), 120, 140, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Ellipse((198, 300,), 120, 140, linestyle='--', linewidth=2, color='w', fill=False))
        ax.add_artist(Ellipse((99, 300,), 120, 140, linestyle='--', linewidth=2, color='w', fill=False))
        ax.annotate("",xy=(260, 400), xycoords='data', xytext=(200, 400), textcoords='data',
                    arrowprops=dict(arrowstyle="->", fc="w", ec="w", linewidth=2))
        ax.annotate("",xy=(200, 470), xycoords='data', xytext=(200, 400), textcoords='data',
                    arrowprops=dict(arrowstyle="->", fc="w", ec="w", linewidth=2))
        
        ax.text(0.47, 0.825, r"$r_{1,2}$", ha='center', va='center', 
                transform=ax.transAxes, color="w", fontsize=38, rotation=0)
        ax.text(0.43, 0.9, r"$r_{3}$", ha='center', va='center', 
                transform=ax.transAxes, color="w", fontsize=38, rotation=0)
        
    ax.imshow(image, cmap='Greys',  interpolation='nearest')
    for item in ax.get_xticklabels():
        item.set_fontsize(28)
    for item in ax.get_yticklabels():
        item.set_fontsize(28)

    ax.set_xlabel(x_label, size=40)
    ax.set_ylabel(y_label, size=40)
    ax.set_xlim(0, im.shape[0]+1)
    ax.set_ylim(0, im.shape[1]+1)
    
fig.canvas.draw()

plt.show()
fig.savefig("image_data/image_views/image_views_results_section_clarified.png", bbox_extra_artists=None, bbox_inches='tight', dpi=300)
#plt.close()
```
