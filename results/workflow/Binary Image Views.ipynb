```python
import matplotlib.pyplot as plt
import numpy as np
import tifffile
%matplotlib inline
```

```python
fig, axarr = plt.subplots(1, 3, figsize=(48,16))
ax1 = axarr[0]
ax2 = axarr[1]
ax3 = axarr[2]

names = ["beadpack", "ketton","bentheimer", "estaillades", "doddington",
               "S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9", "C1", "C2","Berea",
              "Isotropic12", "Isotropic13", "Isotropic14", "Anisotropic121214", "Anisotropic131314"]
direc = "image_data/images/"

for name in names:
    im = tifffile.imread(direc+name+".tif")

    for ax, title, image, x_label, y_label in zip(axarr, [r'$X-Y \ Plane$', r'$X-Z \ Plane$', r'$Y-Z \ Plane$'],
                                                 [im[0, :, :], im[:, 0, :], im[:, :, 0]],
                                                 [r'$x-Axis \ (Voxels)$', r'$x-Axis \ (Voxels)$', r'$y-Axis \ (Voxels)$'],
                                                 [r'$y-Axis \ (Voxels)$', r'$z-Axis \ (Voxels)$', r'$z-Axis \ (Voxels)$']):

        ax.set_title(title, size=45, y=1.05)

        #labels = [item.get_text() for item in ax.get_yticklabels()]
        #print labels
        #ax.set_yticklabels(labels[::-1])

        ax.imshow(image, cmap='Greys',  interpolation='nearest')
        for item in ax.get_xticklabels():
            item.set_fontsize(28)
        for item in ax.get_yticklabels():
            item.set_fontsize(28)

        ax.set_xlabel(x_label, size=40)
        ax.set_ylabel(y_label, size=40)
        ax.set_xlim(0, im.shape[0]+1)
        ax.set_ylim(0, im.shape[1]+1)
    fig.canvas.draw()

    plt.show()
    fig.savefig("image_data/image_views/"+name+"_binary_views.png", bbox_extra_artists=None, bbox_inches='tight', dpi=300)
    plt.close()
```
